```python
import numpy as np
import cv2
```

```python
img = cv2.imread('19.2 capstone_coins.png', cv2.IMREAD_GRAYSCALE)
original_file = cv2.imread('19.2 capstone_coins.png', 1)
img = cv2.GaussianBlur(img, (5,5),0)
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT, 0.9,120,param1=50,param2=27,minRadius=60,maxRadius=120)
print(circles)

circles = np.unit16(np.around(circles))
count = 0
for i in circles[0,:]:
    cv2.circle(original_file,(i[0],i[1]),i[2],(0,255,0),2)
    cv2.circle(original_file,(i[0],i[1]),2,(0,0,255),3)
    cv2.putText(original_file,str(count),(i[0],i[1]), cv2.FONT_HERSHEY_SIMPLEX)

cv2.imshow('Detected coins', original_file)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
